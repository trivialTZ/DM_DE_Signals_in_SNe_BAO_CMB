```python
import corner
import matplotlib.pyplot as plt
import numpy as np
import pickle

samples_sne_bao_ga_obh2 = np.load('data/null_lcdm_BAO_SNe_same_omm_all.npy')

samples_sne_bao_cmb_ga_obh2 = np.load('data/null_lcdm_CMB_SNe_BAO_same_omm_all.npy')
```

```python
samples_sne_bao_ga_obh2
```

```python
import corner
import matplotlib.pyplot as plt
import numpy as np
import pickle
figure1 = corner.corner(samples_sne_bao_ga_obh2,
                        labels=[r"$\Omega_m$", r"$w_0$", r"$w_a$", r"$H_0$", "M"],
                        show_titles=True,
                        truths=[0.3, -1, 0, 70,0],
                        title_kwargs={"fontsize": 14},
                        quantiles=[0.16, 0.5, 0.84],
                        label_kwargs={"fontsize": 14},
                        plot_datapoints=True,
                        plot_density=True,
                        levels=(0.68, 0.95, 0.997))
figure1.suptitle("Null Test: SNe, BAO\nGaussian $\Omega_b h^2$ prior", fontsize=22)
figure1.savefig('corner_null_test_sne_bao_ga_obh2.png', dpi=400)
```

```python

# Generate the second corner plot (SNe, BAO, and CMB)
figure2 = corner.corner(samples_sne_bao_cmb_ga_obh2,
                        labels=[r"$\Omega_m$", r"$w_0$", r"$w_a$", r"$H_0$","M"],
                        show_titles=True,
                        truths=[0.3, -1, 0, 70,0],
                        title_kwargs={"fontsize": 14},
                        quantiles=[0.16, 0.5, 0.84],
                        label_kwargs={"fontsize": 14},
                        plot_datapoints=True,
                        plot_density=True,
                        levels=(0.68, 0.95, 0.997))
figure2.suptitle("Null Test: SNe, BAO, CMB\nGaussian $\Omega_b h^2$ prior", fontsize=22)
figure2.savefig('corner_null_test_sne_bao_cmb_ga_obh2.png', dpi=400)

plt.show()
fig, axes = plt.subplots(1, 2, figsize=(24, 12))

left_img = plt.imread("corner_null_test_sne_bao_ga_obh2.png")
right_img = plt.imread("corner_null_test_sne_bao_cmb_ga_obh2.png")

axes[0].imshow(left_img)
axes[0].axis('off')
axes[1].imshow(right_img)
axes[1].axis('off')

fig.savefig("cornor_null_with_gaussian_ga_obh2.png", dpi=400, bbox_inches='tight')
plt.show()
```
