```python
import numpy as np
import matplotlib.pyplot as plt
import corner


#cmbv1
samples_selectedwith_H0_nocmb = np.load('samples_selectedwith_H0_nocmb.npy')
samples_selected = np.load('samples_selected.npy')

```

```python
plt.rc('font', family='serif', serif='Times New Roman', size=14)
plt.rcParams['axes.labelsize'] = 14

# left panel (without CMB)
figure_left = corner.corner(samples_selectedwith_H0_nocmb, 
                            labels=[r"$\Omega_m$", r"$w_0$", r"$w_a$"],  
                            truths=[np.nan, -1, 0], 
                            show_titles=True, 
                            title_kwargs={"fontsize": 14},
                            quantiles=[0.16, 0.5, 0.84],
                            label_kwargs={"fontsize": 14},
                            plot_datapoints=True,
                            plot_density=True,
                            levels=(0.68, 0.95, 0.997),
                            contour_kwargs={"linewidths": 1.5})
figure_left.savefig("corner_left.png", dpi=300, bbox_inches='tight')

figure_right = corner.corner(samples_selected, 
                             labels=[r"$\Omega_m$", r"$w_0$", r"$w_a$"],  
                             truths=[np.nan, -1, 0], 
                             show_titles=True, 
                             title_kwargs={"fontsize": 14},
                             quantiles=[0.16, 0.5, 0.84],
                             label_kwargs={"fontsize": 14},
                             plot_datapoints=True,
                             plot_density=True,
                             levels=(0.68, 0.95, 0.997),
                             contour_kwargs={"linewidths": 1.5})
figure_right.savefig("corner_right.png", dpi=300, bbox_inches='tight')

fig, axes = plt.subplots(1, 2, figsize=(24, 12))  

left_img = plt.imread("corner_left.png")
right_img = plt.imread("corner_right.png")

axes[0].imshow(left_img)
axes[0].axis('off')  
axes[1].imshow(right_img)
axes[1].axis('off')

#axes[0].text(0.5, -0.1, r"Mock $\Lambda$CDM: SNe ($\Omega_m = 0.35$), BAO ($\Omega_m = 0.30$)", 
            # ha='center', va='center', transform=axes[0].transAxes, fontsize=12)
#axes[1].text(0.5, -0.1, r"Mock $\Lambda$CDM: SNe ($\Omega_m = 0.35$), BAO ($\Omega_m = 0.30$), CMB ($\Omega_m = 0.315$)", 
        #     ha='center', va='center', transform=axes[1].transAxes, fontsize=12)

fig.savefig("cmb_bao_sne_lcdm_to_w0wa_combined_high_res.png", dpi=1200, bbox_inches='tight')
plt.show()
```
